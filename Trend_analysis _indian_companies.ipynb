```python
import pandas as pd
```

```python
url="https://en.wikipedia.org/wiki/List_of_companies_of_India"
tables=pd.read_html(url)
```

```python
df=tables[1]
df
```

```python
df.info()
```

```python
df.columns=df.columns.map(lambda x: x[0]) #reducing from multi-index to single index  
```

```python
df.info()
```

```python
df.head()
```

```python
df['Industry'].value_counts()
```

```python
df['Sector'].value_counts()
```

```python
df['Founded']=pd.to_numeric(df['Founded'],errors='coerce')
df
```

```python
founded_counts=df['Founded'].value_counts().sort_index()
founded_counts
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x=founded_counts.index,y=founded_counts.values,markers='o')
plt.title("Trend Analysis : No of companies founded per year")
plt.xlabel("Year founded ")
plt.ylabel("no. of companies founded")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df,x='Industry',order=df['Industry'].value_counts().index)
plt.title("Trend analysis : Number of companies founded by industry")
plt.xlabel("Industry")
plt.ylabel("Number of companies")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
